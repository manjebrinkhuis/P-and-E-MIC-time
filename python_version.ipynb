# P-and-E-MIC-time

```python
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
```

```python
def make_note(midi):
    return 2**((midi-49) / 12) * 440
```

```python
fs = 44100
secs = 2

c_min = np.array([
    make_note(52),
    make_note(55),
    make_note(59)
])

amp = np.array([
    1,
    .5,
    .8
])

x = np.arange(fs * secs) / fs
X = np.tile(x, c_min.size).reshape((c_min.size, x.size))

Y = np.sin(X.T * (2 * np.pi) * c_min) * amp 

y = Y.sum(axis=1)
y /= y.max() 
y /= 2
y += .5
```

```python
plt.figure(figsize=(20,4))
plt.plot(y)
```

```python
sd.play(np.round(y) - .5)
```
